```python
import torch
import re
from transformers import MT5ForConditionalGeneration, T5Tokenizer
from NewsDBSqlite import NewsDBSqlite

def remove_extra_id(summary):
    return re.sub(r'(<extra_id_\d+>)', '', summary)

def summarize(text, model_name="google/mt5-small"):
    device = torch.device("cuda" if torch.cuda.is_available() else "mps" if torch.backends.mps.is_available() else "cpu")
    print(device)
    
    tokenizer = T5Tokenizer.from_pretrained(model_name)
    model = MT5ForConditionalGeneration.from_pretrained(model_name).to(device)
    prefix = "summarize: "
    if isinstance(text, list):
        input = [prefix + doc for doc in text]
    else:
        input = prefix + text

    inputs = tokenizer(input, return_tensors="pt", max_length=1024, padding=True, truncation=True).to(device)
    summary_ids = model.generate(inputs.input_ids, num_beams=8, max_length=150, early_stopping=True)

    if (isinstance(text, list)):
        result = []
        for i, summary_id in enumerate(summary_ids):
            summary = tokenizer.decode(summary_id, skip_special_tokens=True)
            summary = remove_extra_id(summary)
            result.append(summary)
        summary = result

    else:
        summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
        summary = remove_extra_id(summary)

    return summary

if __name__ == "__main__":
    # https://github.com/google-research/multilingual-t5 mT5-XXL
    newsDB = NewsDBSqlite()
    rows = newsDB.readAll()
    text = []
    count = 0
    for id, times, title, link, source, content in rows:
        count += 1
        # print(content)
        # data = content
        # data = content.replace('\n', '')
        if len(content) < 1024 and len(content) > 512:
            text.append(content)
        if text.__len__() == 5:
            for t in text:
                print(t)
                print('==========================================')
            summary = summarize(text, 'google/mT5-large')
            for s in summary:
                print(s)
                print('==========================================')
            text = []
            print('==========================================')
            print('==========================================')
```
